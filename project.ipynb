```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
f = pd.read_csv(r'C:\Users\KDT-37\Desktop\KDT_7\07.numpy\only_medal.csv')
only_medalDF = pd.DataFrame(f)
```

```python
only_medalDF = only_medalDF.loc[(only_medalDF['Year'] >= 1960) & (only_medalDF['Season'] == "Summer"), :]
```

```python
only_medalDF.head()
```

```python
identify_team_events = pd.pivot_table(only_medalDF ,index = ['Team', 'Year', 'Event'], columns = 'Medal', 
                                      values = 'Medal_Won', aggfunc = 'sum', fill_value = 0).reset_index()
```

```python
identify_team_events = identify_team_events.loc[identify_team_events['Gold'] > 1, :]
print(identify_team_events)
```

```python
team_sports = identify_team_events['Event'].unique()
print(team_sports)
```

```python
# team과 sigle 나누기 
team = only_medalDF['Event'].map(lambda x: x in team_sports)
single = [not i for i in team]

medal = only_medalDF['Medal_Won'] == 1

# team과 medal이 모두 True인 경우에만 1을 반환하고, 그렇지 않으면 0
only_medalDF['Team_Event'] = np.where(team & medal, 1, 0)

# single과 medal이 모두 True인 경우에만 1을 반환하고, 그렇지 않으면 0
only_medalDF['Single_Event'] = np.where(single & medal, 1, 0)

only_medalDF['Event_Result'] = only_medalDF['Single_Event'] + only_medalDF['Team_Event']
```

```python
only_medalDF.drop('Unnamed: 0', axis=1)
```

```python
# medal_total = only_medalDF.groupby(['Year', 'Team', 'Event', 'Medal'])[['Medal_Won', 'Event_Result']].agg('sum').reset_index()
medal_total = only_medalDF.groupby(['Name', 'Sex', 'Year', 'Team', 'Medal', 'GDP'])[['Medal_Won', 'Event_Result']].agg('sum').reset_index()

medal_total['Medal_Won_Corrected'] = medal_total['Medal_Won']/medal_total['Event_Result']
```

```python
medal_total['Medal_Won_Corrected'] = medal_total['Medal_Won_Corrected'].astype(int)
```

```python
# 연도별 각 나라의 선수 수 

# 이름 중복 데이터 제거
year_team = only_medalDF.loc[:, ['Year', 'Team', 'Name', 'Sex']].drop_duplicates()
```

```python
# 'Sex'을 기준으로 'Year'와 'Team'별 선수 수를 집계
year_team_count = pd.pivot_table(year_team, index=['Year','Team'],
                                 columns='Sex',
                                 aggfunc='count').reset_index()
year_team_count
```

```python
# 다중인덱스 제거
year_team_count.columns = year_team_count.columns.get_level_values(0)
year_team_count
```

```python
year_team_count.columns = ['Year', 'Team', 'Female_Athletes', 'Male_Athletes']
```

```python
# 결측치 제거
year_team_count = year_team_count.fillna(0)
```

```python
year_team_count
```

```python
year_team_count['Total_Athletes'] = (year_team_count['Female_Athletes'] + year_team_count['Male_Athletes'])
year_team_count['Total_Athletes'] = year_team_count['Total_Athletes'].astype(int)
```

```python
year_team_count.drop(['Female_Athletes','Male_Athletes'], axis=1, inplace=True)
```

```python
year_team_count
```
